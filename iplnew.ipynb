```python
fig = go.Figure()

fig.add_trace(go.Bar(x=ipl['team1'],y=ipl[ 'second_ings_score'],name='highscore',marker_color='Blue'))

fig.add_trace(go.Bar(x=ipl['team2'],y=ipl['second_ings_score'],name='highscore',marker_color='red'))

fig.update_layout(barmode='group', xaxis_tickangle=50,title='Winners Vs Innings Score')

fig.show()
```

```python

plt.bar(ipl['team1'],ipl['first_ings_score'],color='red',width=0.3,align='edge',
        bottom=40)
plt.xlabel('team1')
plt.xticks(rotation=90)
plt.ylabel('Total Score')
plt.title('Team 1 vs First Inning', fontsize=20)

plt.show()


plt.bar(ipl['team1'],ipl['second_ings_score'],color='pink',width=0.3,align='edge',
        bottom=40)
plt.xlabel('team1')
plt.xticks(rotation=90)
plt.ylabel('Total Score')
plt.title('Team 1 vs Second Inning', fontsize=20)

plt.show()



plt.bar(ipl['team2'],ipl['first_ings_score'],color='green',width=0.3,align='edge',
        bottom=40)
plt.xlabel('team2')
plt.xticks(rotation=90)
plt.ylabel('Total Score')
plt.title('Team 2 vs First Inning', fontsize=20)
plt.show()


plt.bar(ipl['team2'],ipl['second_ings_score'],color='blue',width=0.3,align='edge',
        bottom=40)
plt.xlabel('team2')
plt.xticks(rotation=90)
plt.ylabel('Total Score')
plt.title('Team 2 vs Second Inning', fontsize=20)
plt.show()
```
